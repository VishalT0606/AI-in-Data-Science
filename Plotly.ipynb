data vizualization with plotly

```python
#libraries
import plotly.express as px #express #object_go
import pandas as pd
import matplotlib.pyplot as plt
```

```python
India_data=px.data.gapminder().query("country=='India'")
India_data.head()
```

```python
#bar graph
graph=px.bar(India_data,x='year',y='pop',title='Population of India',labels={'pop':'Population','year':'year'},color='pop',hover_data=['lifeExp','gdpPercap'],text='pop',template='plotly_dark')
graph.update_traces(texttemplate='%{text:.2s}',textposition='inside')
graph.update_layout(uniformtext_minsize=8,uniformtext_mode='hide',xaxis_tickangle = -45,yaxis_type='log')
graph.show()
```

line graph

```python
#India population over the year
graph=px.line(India_data,x='year',y='pop',title='Population of India',labels={'pop':'Population','year':'year'},color='pop',hover_data=['lifeExp','gdpPercap'],text='pop',template='plotly_dark')
graph.show()
```

scatter plot

```python
graph=px.scatter(India_data,x='gdpPercap',y='lifeExp',size='pop',title='Life Exp vs per capita',color='pop',hover_data=['lifeExp','gdpPercap'],text='pop',template='plotly_dark')
graph.update_traces(texttemplate='%{text:.2s}')
graph.show()
```

```python
#pie char
px.pie(India_data,values='pop',names='year',title='Population of India',template='plotly_dark',hole=0.3)
```

```python
#histogram
graph=px.histogram(India_data,x='year',y='pop')
graph.show()
```

```python
#heatmap
graph=px.density_heatmap(India_data,y='pop',x='year')
graph.show()
```

```python
#boxplot
graph=px.box(India_data,y='pop',points="all",color_discrete_sequence=['red'])
graph.update_traces(boxmean=False,jitter=1.0,marker=dict(size=5),opacity=0.3)
graph.show()
```

```python
#violenplot
graph=px.violin(India_data,y='pop')
graph.show()
```

```python
#regressionplot
graph=px.scatter(India_data,x='gdpPercap',y='lifeExp',trendline='ols')
graph.show()
```
