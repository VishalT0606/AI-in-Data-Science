Mention each data point name with the help of loop

```python
# import library
import matplotlib.pyplot as plt

names = ["Vishal", "Manav", "Saurabh", "Ajay", "Vijay"]
marks = [20, 40, 50, 60, 50]

plt.figure(figsize=(8, 5))
plt.plot(names, marks, marker="o", linestyle=":", linewidth=2, color="y")

# loop to label each data point
for i in range(len(names)):
    plt.text(names[i], marks[i], names[i], ha="center", va="bottom", fontsize=9)

plt.xlabel("Name of Students")
plt.ylabel("Marks")
plt.title("Marks of Students")
plt.grid(True)
plt.show()
```

Can we draw to line in a single graph? and add legend in your graph.

yes we can.



```python
import matplotlib.pyplot as plt

names = ["Vishal", "Manav", "Saurabh", "Ajay", "Vijay"]

marks_sem1 = [20, 40, 50, 60, 50]
marks_sem2 = [30, 45, 55, 65, 60]

plt.figure(figsize=(5, 5))

plt.plot(names, marks_sem1, marker="o", linestyle=":", label="Semester 1")
plt.plot(names, marks_sem2, marker="s", linestyle="--", label="Semester 2")

plt.xlabel("Name of Students")
plt.ylabel("Marks")
plt.title("Marks Comparison")
plt.legend()   # added legend in the graph for showwing semester 1 and 2 at the top
plt.grid(True)
plt.show()
```

Draw subplot for line chart

```python
import matplotlib.pyplot as plt

names = ["Vishal", "Manav", "Saurabh", "Ajay", "Vijay"]

marks_sem1 = [20, 40, 50, 60, 50]
marks_sem2 = [30, 45, 55, 65, 60]

plt.figure(figsize=(10, 4))

# Subplot 1
plt.subplot(1, 2, 1)
plt.plot(names, marks_sem1, marker="o")
plt.title("Semester 1 Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.grid(True)

# Subplot 2
plt.subplot(1, 2, 2)
plt.plot(names, marks_sem2, marker="s")
plt.title("Semester 2 Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.grid(True)

plt.tight_layout()
plt.show()
```
