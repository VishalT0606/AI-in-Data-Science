Merge in dataframe


```python
import pandas as pd
# create Df's
df1 =pd.DataFrame({"A":[76,85,8],"B":[67,8,60]})
df2 =pd.DataFrame({"A":[80,851,80],"C":[80,44,5]})

#PERFORM MERGE
pd.merge(df1,df2)
```

```python
#using how
df1 =pd.DataFrame({"A":[76,85,80,6],"B":[67,8,60,4]})
df2 =pd.DataFrame({"A":[76,85,80,5],"C":[80,44,5,2]})

#PERFORM MERGE
pd.merge(df1,df2,on="A",how="outer")

```

```python
#using how
df1 =pd.DataFrame({"A":[76,85,80,6],"B":[67,8,60,4]})
df2 =pd.DataFrame({"A":[76,85,80,5],"C":[80,44,5,2]})

#PERFORM MERGE
pd.merge(df1,df2,on="A",how="inner")
```

```python
#using how
df1 =pd.DataFrame({"A":[76,85,80,6],"B":[67,8,60,4]})
df2 =pd.DataFrame({"A":[76,85,80,5],"C":[80,44,5,2]})

#PERFORM MERGE
pd.merge(df1,df2,on="A",how="right")
```

```python
#using how
df1 =pd.DataFrame({"A":[76,85,80,6],"B":[67,8,60,4]})
df2 =pd.DataFrame({"A":[76,85,80,5],"C":[80,44,5,2]})

#PERFORM MERGE
pd.merge(df1,df2,on="A",how="left")
```

Index


```python
df1 =pd.DataFrame({"A":[76,85,80,6],"B":[67,8,60,4]})
df2 =pd.DataFrame({"A":[76,85,80,6],"B":[80,44,5,2]})

#PERFORM MERGE
pd.merge(df1,df2,left_index=True ,right_index=True,suffixes=("_id","_name"))
```

```python
s1 =pd.Series([76,85,80,6])
s2 =pd.Series([761,851,80,6])
pd.concat([s1,s2])
```

```python
s1 =pd.DataFrame([76,85,80,6] ,[5,8,76,])
s2 =pd.DataFrame([761,851],[8,9,6,6])
pd.concat([s1,s2])
```

```python
df1 =pd.DataFrame({"A":[76,85,80,6],"B":[67,8,60,4]})
df2 =pd.DataFrame({"A":[76,85,80,6],"B":[80,44,5,2]})

pd.concat([df1,df2],axis=1,keys=["df1","df2"])
```

```python
df1 =pd.DataFrame({"A":[76,85,80,6]})
df2 =pd.DataFrame({"A":[76,85,80,6],"B":[80,44,5,2]})

pd.concat([df1,df2],axis=0,keys=["df1","df2"])
```
