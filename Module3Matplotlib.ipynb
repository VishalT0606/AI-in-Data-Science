Data visualization with matplotlib

```python

#import libraries
import matplotlib.pyplot as plt
x=["Vishal", "manav","saurabh","Ajay","Vijay"]
y=[20,40,50,60,50]
plt.figure(figsize=(5,5))

plt.plot(x,y,marker="o",linestyle=":",linewidth=2,color="y")
plt.xlabel("Name of students")
plt.ylabel("Marks")
plt.title("Marks of students")
plt.grid(True)

plt.show
```

```python
import pandas as pd
x={"a":[40,40,60,60,70]}
df = pd.DataFrame(x)
df.index=["Vishal", "manav","saurabh","Ajay","Vijay"]

plt.plot(df["a"],marker="o",linestyle=":",linewidth=2,color="y")
plt.xlabel("Name of students",color="red",fontsize="18")
plt.ylabel("Marks")
plt.title("Marks of students")
plt.grid(True)
```

Histogram

```python
plt.hist(df["a"],bins=10,color="y",edgecolor="b",alpha=1)
plt.xlabel("Name of students",color="red",fontsize="18")
plt.ylabel("Marks")
plt.title("Marks of students")
plt.grid(axis="y")
plt.show()
```

Pie Chat

```python
sub = [ " phy", "chem","math"]
marks = [50,30,20]
plt.pie(marks,labels=sub,autopct="%1.1f%%",startangle=180,explode=[0.5,0,0])
plt.show()
```
