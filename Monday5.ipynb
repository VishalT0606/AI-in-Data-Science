```python
Sales = { "Months":["JAN","FEB","MAR","APR","MAY","JUN","JUL","AUG","SEP","OCT","NOV","DEC"],
           "sales_unit":[20,22,24,26,28,30,32,34,35,36,38,39],
          "revenue_in_lakh":[2,3,4,5,6,6,6,7,7,8,8,9]
}
import matplotlib.pyplot as plt
import pandas as pd
plt.plot(Sales["Months"],Sales["revenue_in_lakh"],marker="o")
plt.xlabel("Months",color="red",fontsize=18)
plt.ylabel("revenue_in_lakh",color="red",fontsize=18)
plt.title("monthly revenue",color="red",fontsize=18)
plt.legend(["revenue_in_lakh"],loc="upper left")
plt.grid('True')
for i,j in zip(Sales["Months"],Sales["revenue_in_lakh"]):
    plt.text(i,j,str(j))
plt.show()
```
