Seaborn Assignment

```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
uploaded = files.upload()
```

```python
df=pd.read_csv('penguins.csv')
df
```

Working With Different libraries

```python
#scatter plot
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=df,hue='sex')
```

```python
# Line plot
sns.lineplot(x='bill_length_mm',y='bill_depth_mm',data=df,hue='sex')
```

```python
#histogram
sns.histplot(df['bill_length_mm'],kde=True)
```

```python
#box plot
sns.boxplot(x='species',y='body_mass_g',data=df)
```

```python
#violin plot
sns.violinplot(x='species',y='body_mass_g',data=df)
```

```python
#bar plot
sns.barplot(x='species',y='body_mass_g',data=df)
```

```python
#count plot
sns.countplot(x='species',data=df)
```

```python
#heatmap
sns.heatmap(df.select_dtypes('number').corr(),annot=True)
```

```python
#pair plot
sns.pairplot(df,hue='species')
```

```python
#regression plot
sns.regplot(x='bill_length_mm',y='bill_depth_mm',data=df)
```
