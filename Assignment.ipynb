Merge 3 Dataframe

```python
import pandas as pd
Df1 = pd.DataFrame([1,2,3,4,5])
Df2 = pd.DataFrame([6,7,8,9,10])
Df3 = pd.DataFrame([11,12,13,14,15])

merge12 = pd.merge(Df1, Df2, left_index=True, right_index=True)
print(merge12)
final_merge = pd.merge(merge12, Df3, left_index=True, right_index=True)
print(final_merge)

```

```python
import pandas as pd

import pandas as pd

Df1 = pd.DataFrame([1, 2, 3, 4, 5])
Df2 = pd.DataFrame([6, 7, 8, 9, 10])
Df3 = pd.DataFrame([11, 12, 13, 14, 15])

Df1.columns = ['A']
Df2.columns = ['B']
Df3.columns = ['C']

# Step 1: concatenate DF1 and DF2
result = pd.concat([Df1, Df2],axis=1, ignore_index=True)

# Step 2: join DF3
final = result.join(Df3)

print(final)




```
